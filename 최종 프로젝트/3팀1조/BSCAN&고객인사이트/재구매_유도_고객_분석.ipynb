###월별 판매금액

```python
# 쿠폰 상태가 'Used' 인 경우와 아닌 경우를 구분하여 계산
def calculate_amount(row):
    if row['쿠폰상태'] == 'Used':
        return row['수량'] * row['평균금액'] * (row['할인율'] / 100) * (1 +row['GST'])
    else:
        return row['수량'] * row['평균금액'] * (1 + row['GST'])

# 각 행에 대한 계산 결과를 구합니다.
cluster['총액'] = cluster.apply(calculate_amount, axis=1)

# 월별로 총액을 합산합니다.
result = cluster.groupby('월')['총액'].sum().reset_index()

# 결과를 선 그래프로 시각화
plt.figure(figsize=(10, 6))
plt.plot(result['월'], result['총액'], marker='o')
plt.title('월별 총액의 변화')
plt.xlabel('월')
plt.ylabel('총액')
plt.xticks(result['월'])
plt.grid(True)
plt.show()
```
