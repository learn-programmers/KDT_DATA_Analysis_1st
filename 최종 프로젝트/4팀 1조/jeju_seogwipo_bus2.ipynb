```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from matplotlib.ticker import FuncFormatter  # 필요한 모듈 가져오기

plt.rc('font', family='NanumMyeongjo') # For Windows
print(plt.rcParams['font.family'])
```

```python
station_df = pd.read_csv('제주특별자치도_버스노선현황.csv', encoding='cp949')

station_df.head()
```

```python
jeju_df = station_df[station_df['광역단체'] == '제주시']
seogwipo_df = station_df[station_df['광역단체'] == '서귀포시']
```

```python
address_lists = []
for i in jeju_df['운행구간']:
    tracks = i.split(", ")
    tmp = []
    for j in range(len(tracks)//2):
        tmp.append(tracks[j*2] + ", " + tracks[j*2+1])
    address_lists.append(tmp)
print(address_lists)
```

```python
combined_list = []
for sublist in address_lists:
    combined_list.extend(sublist)

# 중복되는 값을 찾기
duplicates = set([x for x in combined_list if combined_list.count(x) > 1])

print("중복되는 값들:", duplicates)
```

```python
no_dups = set(combined_list) - duplicates
```

```python
print(no_dups)
```

```python
# 중복되는 값의 개수를 세는 딕셔너리 생성
count_dict = {}
for item in combined_list:
    if item in count_dict:
        count_dict[item] += 1
    else:
        count_dict[item] = 1

# 3개 이상 중복되는 값을 찾기
three_or_more_duplicates = [key for key, value in count_dict.items() if value >= 3]

print("3개 이상 중복되는 값들:", three_or_more_duplicates)
```

```python
no_dups_and_long = {'용담레포츠공원, 제주시', '제주대/월평, 제주시', '제주대학교, 제주시', '제주제주민속오일시장, 제주시', '신제주R, 제주시', '삼무공원, 제주시', '백록초, 제주시', '서중, 제주시', '관덕정, 제주시', '도청, 제주시', '용담, 제주시', '오라3동, 제주시'}
```

```python
no_dups_and_long
```

```python
no_dups_and_long.intersection(no_dups)
```

```python
import folium
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="my_geocoder")

# 맵 생성
map_duplicated = folium.Map(location=[33.5, 126.5], zoom_start=11)

# 중복x 값들에 대해 선과 마커 추가
prev_location = None
for location in no_dups:
    geocode = geolocator.geocode(location)
    if geocode:
        current_location = [geocode.latitude, geocode.longitude]
        # 마커 추가
        if location in no_dups_and_long:
            folium.Marker(current_location, popup=location, icon=folium.Icon(color = 'red')).add_to(map_duplicated)
            print(location)
        else:
            folium.Circle(location = current_location,
                radius = 100,
                popup=location,
                color='red'
            ).add_to(map_duplicated)
        prev_location = current_location
        
for location in three_or_more_duplicates:
    geocode = geolocator.geocode(location)
    if geocode:
        current_location = [geocode.latitude, geocode.longitude]
        # 마커 추가
#         folium.Marker(current_location, popup=location, icon=folium.Icon(color = 'red')).add_to(map_duplicated)
        folium.Circle(location = current_location,
            radius = 100,
            popup=location,
            color='blue'
        ).add_to(map_duplicated)
        prev_location = current_location

map_duplicated.save("제주시_최종.html")
# 지도 출력
map_duplicated
```
