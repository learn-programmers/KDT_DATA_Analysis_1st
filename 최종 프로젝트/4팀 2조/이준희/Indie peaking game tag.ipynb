```python
def tag_scan(use_URL, app_ID):
    # Selenium 옵션 설정
    options = Options()
    options.headless = True

    # Chrome Driver 실행
    service = Service(executable_path=ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=options)
    
    driver.get(use_URL)
    
    game_data = []
    time.sleep(10)  # 페이지 로딩 대기
    
    try:
        name = driver.find_element(By.CLASS_NAME, "apphub_AppName").text.strip()
    except NoSuchElementException:
        name = app_ID
        tag_list = ['a', 'b', 'c', 'd', 'e']
        
        game_data = [name] + tag_list
        print(game_data)

        game_df = pd.DataFrame([{
        '게임 이름': game_data[0],
        'Tag 1': game_data[1],
        'Tag 2': game_data[2],
        'Tag 3': game_data[3],
        'Tag 4': game_data[4],
        'Tag 5': game_data[5]
    }])
    
        return game_df
    
    try:
        tags_list = driver.find_elements(By.CLASS_NAME, "glance_tags")
        tags = tags_list[0].find_elements(By.CLASS_NAME, "app_tag")
        tag_list = [tag.text for tag in tags if tag.text and tag.text != '+'][:5]
    except NoSuchElementException:
        tag_list = ["aaa"]

    game_data = [name] + tag_list
    game_data += [''] * (5 - len(game_data) + 1)  # 게임 이름 고려하여 필요한 만큼 빈 스트링 추가
    print(game_data)

    game_df = pd.DataFrame([{
        '게임 이름': game_data[0],
        'Tag 1': game_data[1],
        'Tag 2': game_data[2],
        'Tag 3': game_data[3],
        'Tag 4': game_data[4],
        'Tag 5': game_data[5]
    }])
    
    driver.quit()
    
    return game_df
```

```python
tags_list = []
for i in app_ID_list:
    use_url = base_url + i
    
    temp = tag_scan(use_url, i)
    tags_list.append(temp)
    
df_combined = pd.concat(tags_list, ignore_index=True)
df_combined
```
