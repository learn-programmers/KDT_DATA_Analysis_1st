```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 데이터 로드

path8 ='/content/The 8th.csv'
path9 = '/content/The 9th.csv'
path10_y = '/content/The 10th(YES).csv'
path10_n = '/content/The 10th(NO).csv'

df8 = pd.read_csv(path8)
df9 = pd.read_csv(path9)
df10_y = pd.read_csv(path10_y)
df10_n = pd.read_csv(path10_n)
```

```python
# 선호하는 결제 방식(Preferred Payment Method), 실제 결제 방식(Payment Method), 구매 카테고리(Category) 분석 그래프

# Accessories 칼럼에서 데이터 추출
accessories_df = df8[df8['Category'] == 'Accessories']

# 'Payment Method'로 그룹화하고 'Mismatched Payment Method' 합산
grouped_df = accessories_df.groupby('Payment Method')['Mismatched Payment Method'].sum().reset_index(name='total_mismatched')

# 히스토그램 그리기
plt.bar(grouped_df['Payment Method'], grouped_df['total_mismatched'])
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.title('Accessories Category')

# 각 막대에 해당하는 값을 막대 위에 표시
for i, value in enumerate(grouped_df['total_mismatched']):
    plt.text(i, value, str(value), ha='center', va='bottom')

plt.show()
```

```python
# Frequency of Purchases로 그룹화하고 Purchase Count 합산
grouped_df = df9.groupby('Frequency of Purchases')['Purchase Count'].sum().reset_index(name='total_purchase_count')

# 막대 그래프 그리기
plt.bar(grouped_df['Frequency of Purchases'], grouped_df['total_purchase_count'])
plt.xlabel('Frequency of Purchases')
plt.ylabel('Purchase Count')
plt.title('Total Purchase Count by Frequency of Purchases')

# X축 레이블 간격 조절
plt.xticks(rotation=45, ha='right', rotation_mode='anchor')

# 각 막대에 해당하는 값을 막대 위에 표시
for i, value in enumerate(grouped_df['total_purchase_count']):
    plt.text(i, value, str(value), ha='center', va='bottom')

plt.tight_layout()  # 레이아웃 조절
plt.show()
```

```python
# 나이대(Age Group) 별로 그룹화하고 각 카테고리(Category)의 Total Count 합산
grouped_df = df10_n.groupby(['Age Group', 'Category'])['Total Count'].sum().reset_index()

# seaborn을 사용하여 막대 그래프 그리기
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='Age Group', y='Total Count', hue='Category', data=grouped_df)
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Total Count by Category and Age Group')
plt.legend(title='Category', loc='upper right', bbox_to_anchor=(1.2, 1))

# 각 막대에 해당하는 값을 막대 위에 표시
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=8, color='black')

# 각 나이대 별로 카테고리를 합산한 값을 나이대 밑에 표시
age_group_totals = df10_n.groupby('Age Group')['Total Count'].sum().reset_index()
for i, value in enumerate(age_group_totals['Total Count']):
    plt.text(i, 0, f'Total: {value}', ha='center', va='bottom')
```

```python
# 나이대(Age Group) 별로 그룹화하고 각 카테고리(Category)의 Avg Previous Purchases 합산
grouped_df = df10_n.groupby(['Age Group', 'Category'])['Avg Previous Purchases'].sum().reset_index()

# seaborn을 사용하여 막대 그래프 그리기
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='Age Group', y='Avg Previous Purchases', hue='Category', data=grouped_df)
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Avg Previous Purchases by Category and Age Group')
plt.legend(title='Category', loc='upper right', bbox_to_anchor=(1.2, 1))

# 각 막대에 해당하는 값을 막대 위에 표시
for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=8, color='black')

# 각 나이대 별로 카테고리를 합산한 값을 나이대 밑에 표시
age_group_totals = df10_n.groupby('Age Group')['Avg Previous Purchases'].sum().reset_index()
for i, value in enumerate(age_group_totals['Avg Previous Purchases']):
    plt.text(i, 0, f'Total: {value}', ha='center', va='bottom')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Category 별로 그룹화하고 Payment Method 및 Preferred Payment Method에 대한 Mismatched Payment Method의 합 계산
grouped_df = df8.groupby(['Category', 'Payment Method'])['Mismatched Payment Method'].sum().reset_index()
grouped_preferred_df = df8.groupby(['Category', 'Preferred Payment Method'])['Mismatched Payment Method'].sum().reset_index()

# seaborn을 사용하여 막대 그래프 그리기 (Payment Method)
plt.figure(figsize=(12, 6))
ax1 = sns.barplot(x='Category', y='Mismatched Payment Method', hue='Payment Method', data=grouped_df)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Mismatched Payment Method by Category and Payment Method')
plt.legend(title='Payment Method', loc='upper right', bbox_to_anchor=(1.2, 1))

# 각 막대에 해당하는 값을 막대 위에 표시
for p in ax1.patches:
    ax1.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=8, color='black')

plt.show()

# seaborn을 사용하여 막대 그래프 그리기 (Preferred Payment Method)
plt.figure(figsize=(12, 6))
ax2 = sns.barplot(x='Category', y='Mismatched Payment Method', hue='Preferred Payment Method', data=grouped_preferred_df)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Mismatched Payment Method by Category and Preferred Payment Method')
plt.legend(title='Preferred Payment Method', loc='upper right', bbox_to_anchor=(1.2, 1))

# 각 막대에 해당하는 값을 막대 위에 표시
for p in ax2.patches:
    ax2.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=8, color='black')

plt.show()
```
