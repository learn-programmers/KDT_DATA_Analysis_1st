```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/malgun.ttf"  # 실제 사용하고자 하는 폰트 파일 경로
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

df = pd.read_csv('BankChurners.csv')
df
```

```python
# 결혼 상태에 따른 수
# Married: 결혼, Single: 솔로, Unknown: 모름, Divorced: 이혼     
df['Marital_Status'].value_counts()
```

```python
# 카드 유지한 고객과 이탈한 고객 수
# Existing Customer: 유지 고객, Attrited Customer: 이탈한 고객 
df['Attrition_Flag'].value_counts()
```

```python
# 결혼 상태에 따른 유지 고객,이탈한 고객 테이블
marital_attrition_ratio = pd.crosstab(df['Marital_Status'], df['Attrition_Flag'])
marital_attrition_ratio
```

```python
# 이탈한 고객 비율 열 추가
marital_attrition_ratio['Attrited_ratio'] = marital_attrition_ratio['Attrited Customer'] / (marital_attrition_ratio['Attrited Customer'] + marital_attrition_ratio['Existing Customer'])
marital_attrition_ratio
```

# 결혼 상태에 따른 이탈 여부 비율 시각화

```python
data = marital_attrition_ratio['Attrited_ratio'].reset_index()
plt.figure(figsize=(8, 5))
sns.barplot(x='Marital_Status', y='Attrited_ratio', data=data)
plt.title('결혼 상태에 따른 이탈 여부 비율')
plt.xlabel('결혼 상태')
plt.ylabel('이탈 여부 비율')
plt.show()
```

# 카이제곱 검정으로 Marital_Status,Attrition_Falg간의 관련성 검정

```python
from scipy.stats import chi2_contingency

# 교차 테이블 생성
contingency_table = pd.crosstab(df['Attrition_Flag'], df['Marital_Status'])

# 카이제곱 검정 실행 ,chi2, p, dof, expected => dof와 expected는 사용하지 않아서 _표기
chi2, p, _, _ = chi2_contingency(contingency_table)

# 결과 출력
print(f"Chi-Square Value: {chi2}")
print(f"P-value: {p}")

# 유의수준 0.05에서의 검정 결과 판단
if p < 0.05:
    print("결과: 귀무가설을 기각합니다. 두 변수 간에는 독립성이 없습니다.")
else:
    print("결과: 귀무가설을 기각하지 못합니다. 두 변수 간에 독립성이 있을 수 있습니다.")
```

* 결론: 결혼상태와 이탈률은 관련이 없다.

# 카드 카테고리별 이탈률 및 시각화

```python
df.groupby('Card_Category')['Attrition_Flag'].value_counts(normalize=True).unstack()
```

```python
# 각 카드 카테고리별 이탈률 계산
Card_Category_Attrition = df.groupby('Card_Category')['Attrition_Flag'].value_counts(normalize=True).unstack().loc[:, 'Attrited Customer']

# 시각화
Card_Category_Attrition.plot(kind='bar', color='skyblue', figsize=(10, 6))
plt.title('이탈률에 따른 카드 카테고리 분포')
plt.xlabel('카드 카테고리')
plt.ylabel('이탈률')
plt.show()
```

# 카드 카테고리별 이탈률 독립성 검정

```python
from scipy.stats import chi2_contingency

# 교차표 생성
card_category_attrition_cross = df.groupby('Card_Category')['Attrition_Flag'].value_counts(normalize=True).unstack()

# 카이제곱 독립성 검정 수행
chi2, p, _, _ = chi2_contingency(card_category_attrition_cross)

# 결과 출력
print(f"Chi-Square Value: {chi2}")
print(f"P-value: {p}")

# 결과 해석
if p < 0.05:
    print("결과: 귀무가설을 기각하여 두 변수 간에는 통계적으로 유의한 관계가 있을 수 있습니다.")
else:
    print("결과: 귀무가설을 기각하지 못하여 두 변수 간에는 통계적으로 유의한 관계가 없을 수 있습니다.")
```

* 결론: 유의한 관계가 없다.
