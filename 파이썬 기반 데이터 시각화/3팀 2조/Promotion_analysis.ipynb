# SRP를 활용한 계산
1. 가격 전략 분석 : SRP = 가격 책정 전략을 이해하는 데 중요한 요소, SRP 실제 판매가격 간의 차이 분석으로 프로모션 효과 평가
2. 수익성 분석 : SRP와 실제 판대 데이터를 비교함으로써, 특정 제품이나 카테고리 수익성 평가 가능
3. 고객 행동 분석 : 고객이 SRP에 비해 높거나 낮은 가격에 제품을 구매하는 경향을 분석, 고객의 가격 인식 및 구매 결정 요인 이해
4. 경쟁 분석 : 시장 내 경쟁 제품의 SRP와 비교, 가격 경쟁력을 평가하고 시장 위치를 이해
5. 프로모션 효과 분석 : SRP 기준으로 할인율을 계산하여, 프로모션 전략의 효과를 분석

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import scipy.stats as stats
from scipy.stats import f_oneway
from statsmodels.formula.api import ols
from sklearn.cluster import KMeans
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv("C:\\Users\\tndud\\Project2\\promo_end.csv")
df.head()
```

```python
# 순이익률 칼럼을 데이터에 추가, store_sales : 고객이 상품이나 서비스에 지불한 금액, store_cost : 상품이나 서비스에 대해 지불한 총 금액
# cost : 프로모션 자체에 들어간 비용
df['net_profit'] = df['store_sales'] - (df['store_cost'] + df['cost']) / df['store_sales']

profitable = df[df['net_profit'] >= 0]

profitable_sorted = profitable.sort_values(by = 'net_profit', ascending= False)

print(profitable_sorted.head())
```

```python
# promotion 별로 그룹화하여 평균 구매액 계산
average_sales_per_promotion = df.groupby('promotion_name')['store_sales'].mean().reset_index()
average_sales_per_promotion_sorted = average_sales_per_promotion.sort_values(by='store_sales', ascending=False)

print(average_sales_per_promotion_sorted.head())
```

### 순이익률 0 이상인 값이 5개밖에 없다 0.0025%제외 전부 적자인 프로모션이라는 소리
전부 Bread 이고 5개 묶음 판매, 프로모션 종류는 다양,

그러면 다른 방식으로 평가를 해야한다.
1. 순이익이아닌 총 매출을 기준으로 프로모션의 성공을 평가, 가장 많은 매출생성은 가장 많은 고객을 끌어드림
2. 마진율 계산 = (store_sale - store_cost) / store_sales, 높은 마진율은 더 효율적인 프로모션
3. 단기적 매출 증대 뿐만아니라 고객 유치 및 장기적 브랜드 가치에 영향, 신규 고객 수, 재구매율, 고객 충성도 지표 => 없기에 패스
4. 순이익은 없어도 매출 증가분이 프로모션 비용에 비해 상대적으로 크다면 장기적으로 성공적일 수 있다.
5. 시장 점유율 변화로도 체크할 수 있다.
6. 다른 경쟁사와의 비교 : 시장 내에서의 프로모션의 상대적 효과 측정 가능

가지고 있는 데이터로 가능한 것 = 2번

```python
# 마진율 계산(실제 판매 수익성 평가)
df['margin_rate'] = (df['store_sales'] - df['store_cost']) / df['store_sales'] * 100


df.head()
```

```python
# 상위 5개 프로모션 이름을 리스트로 저장
top_promotions = ['Two_Day_Sale', 'Price_Slashers', 'Price_Winners', 'Price_Savers', 'Two_for_One']

# bronze 등급인 데이터만 필터링
bronze_data = df[df['member_card'] == 'Bronze']

# bronze 데이터 중에서 상위 5개 프로모션에 해당하는 데이터만 선택
selected_bronze_data = bronze_data[bronze_data['promotion_name'].isin(top_promotions)]

# 원하는 칼럼 리스트 생성
selected_columns = ['promotion_name', 'store_sales', 'store_cost', 'net_profit', 'unit_sales', 'margin_rate', 'food_category', 'sales_country', 'gender', 'avg_yearly_income', 'brand_name', 'media_type', 'total_children', 'cost', 'SRP']

# 원하는 칼럼만 선택하여 그룹화
grouped_data_bronze = selected_bronze_data[selected_columns].groupby('promotion_name').agg({
    'store_sales': 'sum',
    'store_cost' : 'sum',
    'net_profit': 'mean',
    'unit_sales': 'mean',
    'margin_rate' : 'mean',
    'food_category' : lambda x : x.mode()[0],
    'unit_sales' : 'mean',
    'sales_country' : lambda x : x.mode()[0],
    'gender' : lambda x : x.mode()[0],
    'avg_yearly_income' : lambda x : x.mode()[0],
    'brand_name' : lambda x : x.mode()[0],
    'media_type' : lambda x : x.mode()[0],
    'total_children' : lambda x : x.mode()[0],
    'cost' : 'mean',
    'SRP' : 'sum'
}).reset_index()

print(grouped_data_bronze)
```

```python
# media_type 별로 그룹화하여 평균 매출과 표준 오차를 계산합니다.
media_sales = bronze_data.groupby('media_type')['store_sales'].agg(['mean', 'count', 'std'])

# 신뢰 구간을 계산합니다. (95% 신뢰 수준)
confidence_level = 0.95
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# 표준 오차 계산
media_sales['sem'] = media_sales['std'] / np.sqrt(media_sales['count'])

# 신뢰 구간의 하한과 상한 계산
media_sales['ci_lower'] = media_sales['mean'] - z_score * media_sales['sem']
media_sales['ci_upper'] = media_sales['mean'] + z_score * media_sales['sem']

media_sales[['mean', 'ci_lower', 'ci_upper']].sort_values(by='mean', ascending=False)

# 가장 오차가 적은 Daily_Paper_and_Radio 가 제일 좋다
```
