```python
import pandas as pd
```

### 현준님이 전처리한 데이터로 진행

```python
df = pd.read_csv('apart_deal.csv')
df.head()
```

```python
df.info()
```

### 서울특별시 지역코드 = [11]로 시작하여 해당하는 코드들만 남기기

```python
apart = df[df['지역코드'].astype(int).astype(str).str[:2] == '11']
apart
```

### 복사본 생성

```python
apart_1 = apart.copy()
```

### 각 구별 지역코드에 해당하는 지역 맵핑후 새로운 Series로 추가 후 ['지역코드'] 삭제

```python
region_mapping = {
    11110: '서울시 종로구',
    11140: '서울시 중구',
    11170: '서울시 용산구',
    11200: '서울시 성동구',
    11215: '서울시 광진구',
    11230: '서울시 동대문구',
    11260: '서울시 중랑구',
    11290: '서울시 성북구',
    11305: '서울시 강북구',
    11320: '서울시 도봉구',
    11350: '서울시 노원구',
    11380: '서울시 은평구',
    11410: '서울시 서대문구',
    11440: '서울시 마포구',
    11470: '서울시 양천구',
    11500: '서울시 강서구',
    11530: '서울시 구로구',
    11545: '서울시 금천구',
    11560: '서울시 영등포구',
    11590: '서울시 동작구',
    11620: '서울시 관악구',
    11650: '서울시 서초구',
    11680: '서울시 강남구',
    11710: '서울시 송파구',
    11740: '서울시 강동구'
}

apart_1['지역명'] = apart_1['지역코드'].map(region_mapping)
apart_1 = apart_1.drop(columns= ['지역코드'])
apart_1
```

### 컬럼 순서 변경 후 인덱스 번호 초기화

```python
col = ['지역명']+[col for col in apart_1 if col != '지역명']
apart_1 = apart_1[col]
```

```python
apart_1.reset_index(drop=True)
```

### 새로운 csv파일로 저장

```python
apart_1.to_csv('seoul_apart.csv', index = False)
```
