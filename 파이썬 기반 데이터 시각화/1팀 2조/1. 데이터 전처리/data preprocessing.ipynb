```python
import pandas as pd

df = pd.read_csv('seoul_apart.csv')
df.head()
```

### 전용 면적 평수로 변환

```python
df['전용면적'] = (df['전용면적'] * 0.3025).round(1)
df.rename(columns={'전용면적': '평수'}, inplace=True)
df
```

### 거래일 년도로 분류, 면적당 가격 컬럼 추가, 아파트 컬럼 제거, 중복 값 제거

```python
df['거래일'] = df['거래일'].str[:4]
df['평당 가격'] = (df['거래금액'] / df['평수']).round(2)
df = df.drop('아파트',axis=1)
df = df.drop(626502,axis=0)
df.head()
```

### 거래일이 2023인 값 삭제

```python
df = df[df['거래일'] != 2023]
```
